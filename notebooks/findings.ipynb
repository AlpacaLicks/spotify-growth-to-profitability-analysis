# Growth occurred across both premium and ad-supported segments

```python
import pandas as pd
import matplotlib.pyplot as plt

findings = pd.read_csv('../data/raw/spotify_metrics_2018-2025.csv')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(findings['year'], findings['premium_subscribers_millions'], marker='o', label='Premium Subscribers')
plt.plot(findings['year'], findings['ad_supported_maus_millions'], marker='o', label='Ad-Supported MAUs')
plt.title('Spotify Premium vs Ad-Supported Users (2018–2025)')
plt.xlabel('Year')
plt.ylabel('Users (millions)')
plt.xticks(findings['year'])
plt.legend()
plt.grid(True)
plt.savefig("../visuals/findings/spotify_premium_vs_ad_supported_users_2018_2025.png", bbox_inches="tight", dpi=300)
plt.show()
```

# Revenue increased alongside audience growth

```python
plt.figure(figsize=(10, 6))
plt.plot(findings['year'], findings['total_revenue_millions'], marker='o')
plt.title('Spotify Total Revenue (2018–2025)')
plt.xlabel('Year')
plt.ylabel('Total Revenue (EUR millions)')
plt.xticks(findings['year'])
plt.grid(True)
plt.savefig("../visuals/findings/spotify_total_revenue_2018_2025.png", bbox_inches="tight", dpi=300)
plt.show()
```

# Operating income remained inconsistent despite growth

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(findings['year'], findings['maus_millions'], marker='o', label='MAUs')
ax1.set_xlabel('Year')
ax1.set_ylabel('MAUs (millions)')
ax1.set_xticks(findings['year'])
ax1.grid(True)
ax2 = ax1.twinx()
ax2.bar(findings['year'], findings['operating_income_millions'], alpha=0.4, label='Operating Income')
ax2.axhline(0, linewidth=1)
ax2.set_ylabel('Operating Income (EUR millions)')
plt.title('Spotify MAUs vs Operating Income (2018 - 2025)')
lines, labels = ax1.get_legend_handles_labels()
bars, bar_labels = ax2.get_legend_handles_labels()
ax1.legend(lines + bars, labels + bar_labels, loc='upper left')
plt.savefig("../visuals/findings/spotify_maus_vs_operating_income_2018_2025.png", bbox_inches="tight", dpi=300)
plt.show()
```
